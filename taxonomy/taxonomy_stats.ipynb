# Taxonomy Stats

```python
from meso_narratives_revision_2 import mesoNarratives as mesoNarratives_with_stance

total_themes = len(mesoNarratives_with_stance)
meso_lengths = [len(mesos) for mesos in mesoNarratives_with_stance.values()]

all_mesos = [meso for mesos in mesoNarratives_with_stance.values() for meso in mesos]
total_mesos = len(all_mesos)

pro_count = sum(1 for m in all_mesos if len(m) > 1 and m[1] == "pro-migrant")
anti_count = sum(1 for m in all_mesos if len(m) > 1 and m[1] == "anti-migrant")

print(f"Total Narrative Themes: {total_themes}")
print(f"Total Meso Narratives: {total_mesos}")
print(f"Min Meso per Theme: {min(meso_lengths)}")
print(f"Max Meso per Theme: {max(meso_lengths)}")
print(f"Median Meso per Theme: {sorted(meso_lengths)[len(meso_lengths) // 2]}")
print(f"Pro-migrant: {(pro_count / total_mesos) * 100:.1f}%")
print(f"Anti-migrant: {(anti_count / total_mesos) * 100:.1f}%")
```
